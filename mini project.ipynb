```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# define file paths
google_file = 'GOOG.csv'
apple_file = 'AAPL.csv'
amazon_file = 'AMZN.csv'

# read csv files using pandas
google_df =pd.read_csv(r"DS_PT_2/Mini/Mini-1/GOOG.csv")
apple_df  =pd.read_csv(r"DS_PT_2/Mini/Mini-1/AAPL.csv")
amazon_df =pd.read_csv(r"DS_PT_2/Mini/Mini-1/AMZN.csv")

# get column adj close in each file
google_close = google_df['Adj Close']
apple_close = apple_df['Adj Close']
amazon_close = amazon_df['Adj Close']

# check for null values
if google_close.isnull().values.any():
    google_close.dropna(inplace=True)
if apple_close.isnull().values.any():
    apple_close.dropna(inplace=True)
if amazon_close.isnull().values.any():
    amazon_close.dropna(inplace=True)

# sum the close values for each company and store as 3 columns
google_sum = google_close.sum()
apple_sum = apple_close.sum()
amazon_sum = amazon_close.sum()

sum_df = pd.DataFrame({'Google': [google_sum], 'Apple': [apple_sum], 'Amazon': [amazon_sum]})

# find mean, median, and std for each company
google_mean = google_close.mean()
google_median = google_close.median()
google_std = google_close.std()

apple_mean = apple_close.mean()
apple_median = apple_close.median()
apple_std = apple_close.std()

amazon_mean = amazon_close.mean()
amazon_median = amazon_close.median()
amazon_std = amazon_close.std()

# print the results
print("Google:")
print("Mean:", google_mean)
print("Median:", google_median)
print("Standard Deviation:", google_std)

print("\nApple:")
print("Mean:", apple_mean)
print("Median:", apple_median)
print("Standard Deviation:", apple_std)

print("\nAmazon:")
print("Mean:", amazon_mean)
print("Median:", amazon_median)
print("Standard Deviation:", amazon_std)

# plot the data using matplotlib
plt.plot(google_close, label='Google')
plt.plot(apple_close, label='Apple')
plt.plot(amazon_close, label='Amazon')
plt.title('Stock Analysis')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
